```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('/content/Online_Retail.csv', encoding='latin1')
```

DATA CLEANING

*   First, descriptions having spaces needs to be removed.
*   Drop rows that do not have invoice numbers
* Drop rows that do not have invoice numbers
remove credit transactions(those with invoice numbers containg C)

```python
df['Description'] = df['Description'].str.strip() #ensure empty spaces
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)#drop rows that do not have invoice numbers
df['InvoiceNo'] = df['InvoiceNo'].astype('str') #convert IncoiceNo to String
df = df[~df['InvoiceNo'].str.contains('C')]#drop rows with invoice contaning C which means Credit Transactions
```

```python
df.shape
```

```python
print(df.isnull().sum())
```

```python
basket = (df [df['Country'] == 'United Kingdom']#get data from samples which have Country as Germany
          .groupby(['InvoiceNo', 'Description'])['Quantity'] #group them on InvoiceNo and Description based on Quantity
          .sum().unstack().reset_index().fillna(0) #sum the quantity, unstack them, fill 0 with nan values
          .set_index('InvoiceNo'))#set index as Invoice Number
```

```python
basket.head()
```

```python
basket.shape
```

```python
#below function converts all positive valus to 1 and rest to 0
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
#applying function to data using applymap
basket_sets = basket.applymap(encode_units)
```

```python
basket_sets.head()
```

```python
basket_sets.shape
```

```python
#generating frequent item sets
frequent_itemsets = apriori(basket_sets, min_support= 0.07, use_colnames=True)
```

```python
print(frequent_itemsets)
```

The final step is to generate the rules with their corresponding support, confidence and lift.

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules[ (rules['lift'] >=6) &
       (rules['confidence'] >=0.8) ]
```

**CONCLUSION**

In looking at the rules, we see that the green and red alarm clocks are purchased together, also the red paper cups, napkins and plates are purchased together in a manner that is higher than the overall probability would suggest.
